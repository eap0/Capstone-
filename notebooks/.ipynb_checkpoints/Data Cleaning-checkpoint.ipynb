```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mort1 = pd.read_csv('../data/mort1_white_m_55+.txt',sep='\t')#,dtype=[np.str,np.str,np.int,np.str,np.str,np.str,np.str,np.str,np.str,np.str,np.str,np.int,np.int,np.str])
mort2 = pd.read_csv('../data/mort2_white_m_20to54.txt',sep='\t')
mort3 = pd.read_csv('../data/mort3_white_m_0to19.txt',sep='\t')
mort4 = pd.read_csv('../data/mort4_white_fm_ns.txt',sep='\t')
mort5 = pd.read_csv('../data/mort5_white_f_55+.txt',sep='\t')
```

```python
mort = pd.concat([mort1,mort2,mort3,mort4,mort5])
mort
```

```python
mort = mort.drop(['Notes','County Code','Age Group','Gender','Race Code','Hispanic Origin Code'],axis=1)
mort 
```

```python
unreliable = []
states = []
rates = []
deaths = []
pop = []
for i in range(len(mort)):
    name = mort[i:i+1].County.to_string()
    states.append(name.split(', ')[-1])
    rate = mort[i:i+1]['Crude Rate'].to_string()
    if 'Unreliable' in rate:
        rates.append(float(mort[i:i+1]['Crude Rate'].to_string().split(' ')[-2]))
        unreliable.append(True)
    elif 'Not Applicable' in mort[i:i+1]['Crude Rate'].to_string(): 
        rates.append(np.nan)
        unreliable.append(True)
    else:
        rates.append(float(mort[i:i+1]['Crude Rate'].to_string().split(' ')[-1]))
        unreliable.append(False)
    if 'Suppressed' in mort[i:i+1]['Deaths'].to_string(): 
        deaths.append(np.nan)
    else:
        deaths.append(int(mort.Deaths[i:i+1].to_string().split(' ')[-1]))
    if 'Not Applicable' in mort[i:i+1]['Population'].to_string(): 
        pop.append(np.nan)
    else:
        pop.append(int(mort[i:i+1]['Population'].to_string().split(' ')[-1]))
mort['State'] = states
mort['Crude Rate'] = rates
mort['Unreliable'] = unreliable
mort['Deaths'] = deaths
mort['Population'] = pop
```

```python
mort['log_pop'] = np.log10(mort['Population'])
mort['log_deaths'] = np.log10(mort['Deaths'])
```

```python
plt.figure(figsize=(8,6))
plt.ylabel('$log_{10}$(Deaths)',fontsize=16)
plt.xlabel('$log_{10}$(Population)',fontsize=16)
plt.scatter(mort['log_pop'],mort['log_deaths'],alpha=.3,s=1)
```

```python
sns.lmplot(data=mort,x='log_pop',y='log_deaths',hue='Age Group Code',height=10,aspect=1.3,palette='rocket')

plt.show()
```
